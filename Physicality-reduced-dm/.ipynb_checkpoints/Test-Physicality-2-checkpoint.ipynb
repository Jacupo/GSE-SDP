```julia
using LinearAlgebra

Pauli = Dict('I'=>[1 0; 0 1], 'X'=>[0 1; 1 0], 'Y'=>[0 -im; im 0], 'Z'=>[1 0; 0 -1])
I = [1 0; 0 1]
function permutationes(vec)
    elements = ["x","y","z"];
    elements2 = ["X","Y","Z"];
    permu = [];
    vecbackup = deepcopy(vec);
    for i=1:3
        for j=1:3
            if j!=i 
                vec = replace(vec, elements[1]=>elements2[i])
                vec = replace(vec,elements[2]=>elements2[j])
                vec = replace(vec,elements[3]=>elements2[3-mod(i+j,3)])
                push!(permu, vec)
                vec = deepcopy(vecbackup)
            end
        end
    end
    
    permu = unique!(permu);
    #println(permu)
    return permu
end


function strtovec(perm)
    #Pauli = Dict("I"=>[1 0; 0 1], "X"=>[0 1; 1 0], "Y"=>[0 -im; im 0], "Z"=>[1 0; 0 -1])
    operator = zeros(Float64, 2^length(perm[1]),2^length(perm[1]))
    for s in perm
        O = get(Pauli,s[1], NaN)
        for x in s[2:end]
            O = kron(O,get(Pauli,x,NaN))
        end
        operator += O
    end

    return operator
end

function term(t)
   return strtovec(permutationes(t))
end

function TEN(opsvec)
    N = size(opsvec,1)

    O = opsvec[1]
    for i=2:N
        O = kron(O,opsvec[i])
    end

    return O
end


function trace(ρ)
    return sum(diag(ρ))
end

function negativity(ρ)
    N = 0
    for λ in eigvals(ρ)
        N += abs(λ)-λ
    end
    return N/2.
end
```
